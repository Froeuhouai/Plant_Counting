```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN, KMeans
import pandas as pd
```

```python
img = Image.open(
    "/home/fort/Documents/APT 3A/Cours/Ekinocs/output_otsu/Output/Session_1/Otsu/OTSU_screen_1920x1080_11_25.jpg"
)
plt.imshow(img)
plt.show()
```

```python
img_array = np.array(img)
X = []
Y = []
mat_coord = np.argwhere(img_array[:, :, 0] == 255)# contient les coordonnées des points blancs correspondants aux plantes
X, Y = np.where(img_array[:, :, 0] == 255)
```

```python
mat_clustered = DBSCAN(eps=70, min_samples=100).fit(mat_coord)
```

```python
fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(111)
scatter = ax.scatter(Y, X, c=mat_clustered.labels_, s=1,cmap='Paired')
plt.imshow(img)
#plt.show()
#fig.show()
#fig.savefig("/home/fort/Bureau/lineaire_cluster.png")
```

```python
dataframe_coord = pd.DataFrame(mat_coord)
dataframe_coord = dataframe_coord.rename(columns={0:'X',1:'Y'})
label = pd.DataFrame(mat_clustered.labels_)
label = label.rename(columns={0:'label'})
dataframe_coord = pd.concat([dataframe_coord,label],axis=1)# dataframe qui regroupe pour chaque pixel blanc sont cluster auquel il appartient
```

```python
rang_1 = dataframe_coord[dataframe_coord["label"]==1] # On peut extraire un rang parmis ceux détecté précédement
```

```python
#rang_1_clustered = DBSCAN(eps=10, min_samples=200).fit(rang_1[["X","Y"]])
rang_1_clustered = KMeans(n_clusters=15,random_state=0).fit(rang_1[["X","Y"]])
```

```python
dataframe_rang_1 = rang_1[['X','Y']].reset_index(drop=True, inplace=False)
label_rang_1 = pd.DataFrame(rang_1_clustered.labels_)
label_rang_1 = label_rang_1.rename(columns={0:'label'})
dataframe_rang_1 = pd.concat([dataframe_rang_1,label_rang_1],axis=1)
```

```python
# Calcul des centre des cluster en faisant la moyenne des distance des points pour le résultat de DBSCAN
liste_label = dataframe_rang_1.label.unique().tolist()
X_center = []
Y_center = []
for lab in liste_label:
    dataframe_rang_label = dataframe_rang_1[dataframe_rang_1.label == lab]
    X_mean = dataframe_rang_label.X.mean()
    Y_mean = dataframe_rang_label.Y.mean()
    X_center.append(X_mean)
    Y_center.append(Y_mean)
    
```

```python

fig = plt.figure(figsize=(8,10))
ax = fig.add_subplot(111)
scatter = ax.scatter(rang_1["X"], rang_1["Y"], c=rang_1_clustered.labels_, s=1)
#scatter = ax.scatter(X_center, Y_center, s=50)
scatter = ax.scatter([rang_1_clustered.cluster_centers_[k][0] for k in range(10)],[rang_1_clustered.cluster_centers_[k][1] for k in range(10)],c='red')
#plt.show()
#fig.show()
```
